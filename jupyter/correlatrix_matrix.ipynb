```python
import xarray as xr
era5_variables = xr.open_dataset('../download.nc')
era5_variables_mean = era5_variables.mean(['latitude', 'longitude'])
era5_variables_mean
```

```python
glofas_variable = xr.open_mfdataset('../glofas_sample/*.nc')
glofas_variable_mean = glofas_variable.mean(['lat','lon'])
```

```python
era5_transformed = era5_variables_mean.to_array(dim='features').T
glofas_transformed = glofas_variable_mean.to_array(dim='features').T
    
glofas_transformed
```

```python
era5_transformed
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def generate_heatmap(X, y, descr='description'):
    df = pd.DataFrame(data=X.values, columns=X.features.values, index=X.time.values)
    df['dis24'] = y.values
    print(df)

    plt.figure(figsize=(25,25))
    cor = df.corr()
    sns.heatmap(cor, annot=True, cmap=plt.cm.Reds)
    plt.title("Correlation Matrix of available variables")
    plt.savefig('../images/correlation_matrix/correlation_matrix1.png')
    plt.show()
    cor_predictand = abs(cor['dis24'])
    feature_importance = cor_predictand[cor_predictand > 0.2]
#     print(descr)
#     print(feature_importance)
    return feature_importance
```

```python
important_features = []

timeinit, timeend = '01-01-2007', '12-01-2007'

ft = generate_heatmap(X=era5_transformed, y=glofas_transformed, descr=f'{timeinit} to {timeend}')
important_features.append(ft)
```
